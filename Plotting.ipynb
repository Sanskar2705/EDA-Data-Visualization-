```python
# from a pandas dataframe
batsman= pd.read_csv('C:/Users/91798/Plotting Using Matplotlib/sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# legend --> location
plt.plot(batsman['index'], batsman['V Kohli'], color= 'green', linestyle= 'dotted', linewidth= 3, label= 'VK')
plt.plot(batsman['index'], batsman['RG Sharma'], color= '#241444', linestyle= 'dashed',  linewidth= 2, label= 'Rohit')
plt.title('R Sharma vs VK career comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
```

# Bar Plot

Bivariate Analysis

Numerical  vs Categorical

Use case -  Aggregate analysis of groups

```python
# Simple bar chart
children = [10, 20, 45, 15, 25]
colors= ['red', 'green', 'blue','yellow', 'pink',]
plt.bar(colors, children, color='black')
```

```python
# horizontal bar chart --> genrally we use it when more no. of categories
plt.barh(colors, children, color='black')
```

```python
# color and label
df= pd.read_csv('C:/Users/91798/Plotting Using Matplotlib/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'] )
```

```python
plt.bar(np.arange(df.shape[0])-0.1 , df['2015'] , width= 0.1, color= 'magenta')
plt.bar(np.arange(df.shape[0]) , df['2016'] , width= 0.1)
plt.bar(np.arange(df.shape[0])+0.1 , df['2017'] , width= 0.1)

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Problem
children = [10, 20, 45, 15, 25]
colors= ['red red red red red', 'green green green', 'blue blue','yellow yellow yelow yellow', 'pink pink pink pink pink',]
plt.bar(colors, children, color='black')
```

```python
# solution
# Problem
children = [10, 20, 45, 15, 25]
colors= ['red red red red red', 'green green green', 'blue blue','yellow yellow yelow yellow', 'pink pink pink pink pink',]
plt.bar(colors, children, color='black')
plt.xticks(rotation= 'vertical')
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2017'], label= '2017')
plt.bar(df['batsman'], df['2016'], bottom= df['2017'], label='2016' )
plt.bar(df['batsman'], df['2015'],  bottom= df['2016']+ df['2017'], label='2015')

plt.legend()
plt.show()
```

# Pie Chart
Univariate/Bivariate analysis

Categorical vs Numerical

Use case-> To find contribution on a standard scale


```python
# simple data
data = [23 , 45, 100, 20, 40]
subjects= ['eng', 'hindi', 'maths', 'farsi', 'sanskarit']
plt.pie(data, labels= subjects)
plt.show()
```

```python
# dataset
df= pd.read_csv('C:/Users/91798/Plotting Using Matplotlib/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percatage and colors
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct='%0.1f%%',colors=['blue', 'cyan', 'yellow', 'red', 'pink', 'brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct='%0.1f%%', explode=[0.1,0,0,0,0,0])
plt.show()
```
